```python
import time
import random
import matplotlib.pyplot as plt
def binary_search(arr, low, high, target):
    if high >= low:
        mid = (high + low) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] > target:
            return binary_search(arr, low, mid - 1, target)
        else:
            return binary_search(arr, mid + 1, high, target)
    else:
        return -1
def measure_time(n, target):
    arr = sorted([random.randint(1, 100000) for _ in range(n)])
    start_time = time.time()
    binary_search(arr, 0, n - 1, target)
    end_time = time.time()
    return end_time - start_time
sizes = [100, 500, 1000, 5000, 10000, 20000, 50000]
times = []
for n in sizes:
    target = random.randint(1, 100000)
    time_taken = measure_time(n, target)
    times.append(time_taken)
plt.plot(sizes, times, marker='o', linestyle='-', color='b')
plt.xlabel('List Size (n)')
plt.ylabel('Time Taken (seconds)')
plt.title('Binary Search: Time Taken vs List Size')
plt.grid(True)
plt.show()
```
