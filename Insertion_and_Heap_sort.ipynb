```python
import random
import time
import matplotlib.pyplot as plt
def insertion_sort(arr):
    n = len(arr)
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
def heapify(arr, n, i):
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2
    if left < n and arr[i] < arr[left]:
        largest = left
    if right < n and arr[largest] < arr[right]:
        largest = right
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)
def heap_sort(arr):
    n = len(arr)
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)
    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        heapify(arr, i, 0)
def measure_time(sort_func, arr):
    start_time = time.time()
    sort_func(arr)
    end_time = time.time()
    return end_time - start_time
def run_experiment():
    sizes = [100, 500, 1000, 5000, 10000]
    insertion_sort_times = []
    heap_sort_times = []
    for n in sizes:
        arr_insertion = random.sample(range(1, 100000), n)
        arr_heap = arr_insertion.copy()
        insertion_time = measure_time(insertion_sort, arr_insertion)
        insertion_sort_times.append(insertion_time)
        heap_time = measure_time(heap_sort, arr_heap)
        heap_sort_times.append(heap_time)
    plt.figure(figsize=(10, 6))
    plt.plot(sizes, insertion_sort_times, label="Insertion Sort", marker='o')
    plt.plot(sizes, heap_sort_times, label="Heap Sort", marker='s')
    plt.xlabel("Number of Elements (n)")
    plt.ylabel("Time Taken (seconds)")
    plt.title("Time vs Number of Elements for Insertion Sort and Heap Sort")
    plt.legend()
    plt.grid(True)
    plt.show()
run_experiment()
```
